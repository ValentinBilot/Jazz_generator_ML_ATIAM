# Cellule d'entraînement
Il faut que le modèle existe pour pouvoir l'entraîner, afin de ne pas écraser par erreur des modèles entraînés, un autre fichier est (sera) disponible pour créer le modèle en question avant de l'entaîner

## Choix des paramètres:

```python
params = {
        'print_every': 1,
        'plot_every': 1,
        'optimizer': "Adam",
        'lossFunction': "CrossEntropy",
        'model_type': "gru",
        'alphabet': 'a0',
        'sequence_lenght': 16, 
        'using_cuda': True,
        'batch_size': 128,
        'shuffle': True, 
        'num_workers': 6,
        'hidden_size': 512,
        'num_layers': 2,
        'dropout': 0.1,
        'learning_rate': 1e-4, 
        'epochs': 10}
```

## Import des trucs utiles

```python
from utilities import NetworksUtil
from utilities.NetworksUtil import TrainModel
from utilities.NetworksUtil import TestModel
from utilities.modelClasses import MYGRU
import torch
dropout = params['dropout']
model_type = params['model_type']
num_layers = params['num_layers']
hidden_size = params['hidden_size']
alphabet = params['alphabet']
sequence_lenght = params['sequence_lenght']
dropoutstr = str(dropout).replace('.',',')
model_string = "models/"+model_type+str(num_layers)+"layers"+str(hidden_size)+"blocks"+alphabet+"alphabet"+str(sequence_lenght)+"lenSeq"+dropoutstr+"dropout.pt"
import torch
model = torch.load(model_string)
```

# Entraînement

```python
TrainModel(model, **params)
```

# Sauvegarder le modèle

```python
torch.save(model, model_string)
```

 La fonction TestModel renvoie la moyenne de la loss et la moyenne de  l'accuracy
